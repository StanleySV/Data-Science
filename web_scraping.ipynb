```python
from bs4 import BeautifulSoup 
import requests 
import pandas as pd
```

```python
req = requests.get("https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1")  # URL of the website which you want to scrape
content = req.content # Get the content
```

```python
soup = BeautifulSoup(content,'html.parser')
```

```python
desc = soup.find_all('div' , class_='_4rR01T') 
prod_desc = []  #the list will have the text content of all the description
for i in range(len(desc)):
    prod_desc.append(desc[i].text) # We can even access the child tags with dot access.
prod_desc
```

```python
commonclass = soup.find_all('li',class_='rgWa7D') # We observe that the classnames for the different specifications are under one div.So we need to apply some method to extract the different features.
 
# Create empty lists for the features

ram=[]
display=[]
camera=[]
storage=[]


for i in range(0,len(commonclass)):
    p=commonclass[i].text # Extracting the text from the tags
    if("RAM" in p): # If RAM is present in the text then append it to the ram list. Similarly do this for the other features as well
        ram.append(p)
    elif("Display" in p or "inch" in p): 
        display.append(p)
    elif("Camera" in p):
        camera.append(p)
    elif("Battery" in p):
        storage.append(p)
```

```python
print(len(storage))
print(len(ram))
print(len(camera))
print(len(display))
print(len(prod_desc))
```

```python
price = soup.find_all('div',class_='_30jeq3 _1_WHN1') # Extracting price of each laptop from the website
prices = []
for i in range(len(price)):
    prices.append(price[i].text)
prices
```

```python
print(len(prices))
```

```python
df = {'Product':prod_desc,'RAM/ROM':ram,'Camera':camera,'Price':prices}
df1 = pd.DataFrame(data = df)
df1
```
